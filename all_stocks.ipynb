```python
from time import sleep
import requests
```

```python
pno = 0
companies = {}
done = False
while not done:
    pno += 1
    print(f'Currently on Page {pno}', end=' ')
    url = f'https://www.screener.in/screens/71064/all-stocks/?page={pno}'
    resp = requests.get(url)
    while resp.status_code != 200:
        resp = requests.get(url)
        if resp.status_code != 200:
            print('Retrying', end=' ')
            sleep(3)
    print(resp.status_code, end=' ')
    soup = BeautifulSoup(resp.text, 'html.parser')
    rows = soup.find_all('tr')
    print(len(rows))
    for i in rows:
        if i.find('td'):
            com = i.findAll('td')[1]
            if com.a['href'] in companies:
                done = True
                break
            companies[com.a['href']] = com.text.strip()
    
```

```python
len(companies)
```

```python
companies
with open('companies.txt', 'w') as f:
    for i in companies:
        f.write(f'{i}\t{companies[i]}\n')
```

```python
read_companies = {}
with open('companies.txt') as f:
    for i in f:
        i = i.strip().split('\t')
        read_companies[i[0]] = i[1]
```
